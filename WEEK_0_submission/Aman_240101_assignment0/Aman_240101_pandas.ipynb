Question 40

```python
taxis_one_day = taxis[(taxis['pickup_time'] >= '2019-01-23 00:06:00') & (taxis['pickup_time'] < '2020-01-24 00:00:00')]
```

```python
taxis_one_day
```

Question 41

```python
df_means = taxis_one_day.resample('H', on='pickup_time').mean(numeric_only=True)
```

```python
df_means
```

Question 42

```python
import matplotlib.pyplot as plt
```

```python

plt.plot(df_means.index, df_means['distance'])
plt.xlabel('Pickup Time (Daily)')
plt.ylabel('Average Distance')
plt.title('Average  Distance Over Time')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
